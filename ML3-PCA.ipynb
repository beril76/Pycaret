##### Decompostion has different packages. PCA is one of them. 

```python
import sklearn.decomposition as de
```

```python
image_pca=de.PCA()
```

```python
ht=image_pca.fit_transform(black_image)
```

```python
image_pca.explained_variance_ratio_.shape
```

```python
image_pca.explained_variance_ratio_[0:3]
```

###### Here 68% of the data is explained in the first component. 9% by the second component and so on .

```python
image_pca.explained_variance_ratio_[0:10].cumsum()
```

```python
plt.plot(image_pca.explained_variance_ratio_.cumsum())
```

```python
plt.plot(image_pca.explained_variance_ratio_[0:100].cumsum())
```

```python
image_pca_20=de.PCA(n_components=20)
```

```python
ht=image_pca_20.fit_transform(black_image)
```

```python
ht.shape
```

```python
image_pca_20.inverse_transform(ht).shape
```

```python
plt.imshow(image_pca_20.inverse_transform(ht),cmap='gray')
```

##### In the case of 10 components

```python
image_pca_10=de.PCA(n_components=10)
```

```python
ht=image_pca_10.fit_transform(black_image)
```

```python
plt.imshow(image_pca_10.inverse_transform(ht),cmap='gray')
```

```python
plt.imshow(ht,cmap='gray')
```

##### For 60 components

```python
image_pca_60=de.PCA(n_components=60)
ht=image_pca_60.fit_transform(black_image)
plt.imshow(image_pca_60.inverse_transform(ht),cmap='gray')
```

###### In PCA, information is not lost. In the case of an image, 5% of the pixels will be removed here (8 lakh pixels here) . PCA can be done when there is very large amounts of data . Once the dimensionality is reduced, computation becomes easier and faster as well .
